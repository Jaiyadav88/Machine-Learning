```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
list1=[20,30,40,50,60]
list2=['a','b','c','d','e']
plt.figure()
plt.title('Random Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.bar(list2,list1,color='cyan',edgecolor='blue',width=0.6)
plt.show()
```

```python
complete=[[1,2,3,4],[7,8,9,10],[9,10,17,10],[1,17,18,20]]
country=['a','b','c','d']
plt.figure()
for i in complete:
    plt.title('Dengue Cases')
    plt.xlabel('cases')
    plt.bar(country,i,color='brown',edgecolor='pink')
    #plt.barh() for horizontal bars
    plt.show()
```

## PIE CHART

```python
label=['a','b','c','d','e']
comp=[85,51,101,20,30]
myexplode=[0,0,0.1
           ,0,0]
color=['cyan','blue','yellow','orange','green']
plt.title('Pie Charts')
plt.pie(comp,labels=label,explode=myexplode,colors=color,shadow=True,startangle=180)
plt.legend(color)
#explode function separates a particular value from a pie chart
plt.show()
```

## Scatter Plot

```python
mat1=np.random.rand(100)
mat2=np.random.rand(100)
plt.figure(figsize=(6,5))
plt.title('Random VS Random')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(mat1,mat2,marker='o',color='red')
plt.show()
```
